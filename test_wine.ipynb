# Classification

```python
from sklearn.datasets import load_wine
import pandas as pd
wine = load_wine()
df = pd.DataFrame(wine["data"], columns=wine["feature_names"]) 
df
```

```python
wine
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(wine["data"], 
                                                    wine["target"], 
                                                    test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
```

```python
import graphviz
from sklearn.tree import export_graphviz
g = export_graphviz(clf, feature_names=wine["feature_names"],
                    class_names=wine["target_names"], filled=True)
result = graphviz.Source(g)
result.render("wine")
result
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test)
print("æ­£ç¢ºçŽ‡", accuracy_score(pre, y_test) * 100, "%")
```

# Regression

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=2)
reg.fit(x_train, y_train)
```

```python
import graphviz
from sklearn.tree import export_graphviz
g = export_graphviz(reg, feature_names=wine["feature_names"], filled=True)
result = graphviz.Source(g)
result.render("wine")
result
```

```python
from sklearn.metrics import r2_score
pre = reg.predict(x_test)
print("r2 score:", r2_score(y_test, pre))
```
